# Data Visualization

```python
#gender distribution in percentage

gender_count = df['gender'].value_counts()

plt.pie(
    gender_count,
    labels = gender_count.index,
    autopct ='%1.1f%%',
    shadow = True       
)

plt.show()
```

```python
#Insight--- male and female almost same, just 0.1 percent male ahead from female, others 2 %
```

```python
#Gender distributionaa
sns.countplot(
    data = df,
    x='gender',
    hue = 'admission_status'
)
```

```python
#Insight----Number of Male and Female students in college admission same and also in admission status while other in less number
```

```python
sns.histplot(data =df,
             x="high_school_gpa",
             hue = "admission_status",
             bins =20,
             kde =True
            )
```

```python
#Insight---Students with higher high GPAs are more likely to be admitted
```

```python
#act score distibution 
sns.histplot(
    data=df,
    x="act_score",
    hue="admission_status",
    bins=30,
    kde= True
)
```

```python
#leadership distribution
sns.countplot(
    data = df,x='leadership_positions', 
    hue='admission_status'
    
)
plt.show
```

```python
#Insight----More likely leadership involvement to be admitted
```

```python
# volunteer hours distribution
sns.histplot(data=df, x="volunteer_hours",hue ='admission_status', bins=50)
```

```python
sns.histplot(data=df, x="social_media_hours",hue ='admission_status', bins=20)
```

```python
# volunteer vs social media hours

sns.scatterplot(x = df['volunteer_hours'],
            y = df['social_media_hours'],
            hue =df["admission_status"]
           )

```

```python
#some column's distribution together

num_cols = ['high_school_gpa','sat_score','act_score','attendance_rate','ap_courses']

for col in num_cols:
    
    plt.figure(figsize=(6,4))
    sns.histplot(data=df, x=col,bins =20, kde=True)
    plt.title(f"Distribution of {col}")
    plt.tight_layout()
    plt.show()
```
