# 📂 Dataset Overview

This dataset contains detailed information about startups that appeared on **US Shark Tank**, including season details, startup profiles, entrepreneurs, investment requests, deal outcomes, individual shark investments, and shark participation.

---

## 📺 Season & Episode Information
These features identify when and where each pitch took place.

- **Season Number**
- **Episode Number**
- **Pitch Number**
- **Season Start**
- **Season End**
- **Original Air Date**

---

## 🚀 Startup Information
These features describe the startup and its business.

- **Startup Name**
- **Industry**
- **Business Description**
- **Company Website**

---

## 👥 Entrepreneur Information
These features describe the founders presenting the startup.

- **Pitchers Gender**
- **Pitchers Average Age**
- **Pitchers City**
- **Pitchers State**
- **Entrepreneur Names**
- **Multiple Entrepreneurs**

---

## 💼 Business & Investment Request
These features summarize the startup's business metrics and funding request.

- **US Viewership**
- **Original Ask Amount**
- **Original Offered Equity**
- **Valuation Requested**

---

## 🤝 Deal Information
These features describe the final outcome of the pitch.

- **Got Deal**
- **Total Deal Amount**
- **Total Deal Equity**
- **Deal Valuation**
- **Number of Sharks in Deal**
- **Investment Amount Per Shark**
- **Equity Per Shark**
- **Royalty Deal**
- **Advisory Shares Equity**
- **Loan**
- **Deal Has Conditions**

---

## 🦈 Individual Shark Investments
These features record the investment amount and equity offered by each shark.

### Barbara Corcoran
- Barbara Corcoran Investment Amount
- Barbara Corcoran Investment Equity

### Mark Cuban
- Mark Cuban Investment Amount
- Mark Cuban Investment Equity

### Lori Greiner
- Lori Greiner Investment Amount
- Lori Greiner Investment Equity

### Robert Herjavec
- Robert Herjavec Investment Amount
- Robert Herjavec Investment Equity

### Daymond John
- Daymond John Investment Amount
- Daymond John Investment Equity

### Kevin O'Leary
- Kevin O'Leary Investment Amount
- Kevin O'Leary Investment Equity

### Guest Shark
- Guest Name
- Guest Investment Amount
- Guest Investment Equity

---

## 🪑 Shark Presence
These binary features indicate whether each shark was present during the pitch.

- **Barbara Corcoran Present**
- **Mark Cuban Present**
- **Lori Greiner Present**
- **Robert Herjavec Present**
- **Daymond John Present**

# Handling Missing Values

```python
#X_train.select_dtypes(include='number').skew()
```

```python
 #nunique
#Pitchers Gender                          3
#Pitchers Average Age                     3
#Multiple Entrepreneurs                   2
```

```python
from sklearn.impute import SimpleImputer

mode_imputer = SimpleImputer(strategy='most_frequent')

cols = [
    "Pitchers Gender",                     
    'Pitchers Average Age',                                   
    'Multiple Entrepreneurs'
    
]

mode_imputer.fit(X_train[cols])

X_train[cols] = mode_imputer.transform(X_train[cols])
X_test[cols] = mode_imputer.transform(X_test[cols])
```

# Encoding

```python
print(X_train["Industry"].unique())
print(X_train['Pitchers Gender'].unique())
print(X_train['Pitchers Average Age'].unique())
```

```python
# Ordinal encoding
from sklearn.preprocessing import OrdinalEncoder, OneHotEncoder

age_encoder = OrdinalEncoder(categories=[['Young', 'Middle', 'Old']])

age_encoder.fit(X_train[['Pitchers Average Age']])

# transform

X_train['encoded_Pitchers Average Age'] = age_encoder.transform(
    X_train[['Pitchers Average Age']]
).ravel()

X_test['encoded_Pitchers Average Age'] = age_encoder.transform(
    X_test[['Pitchers Average Age']]
).ravel()

X_test
```

```python
X_train
```

```python

#One- hot - Encoding 

pitchers_gender_ohe = OneHotEncoder(sparse_output=False).set_output(transform='pandas')

pitchers_gender_ohe.fit(X_train[['Pitchers Gender']])

encoded_df = pitchers_gender_ohe.transform(X_train[['Pitchers Gender']])

X_train = pd.concat([X_train, encoded_df], axis=1)
X_train
```

```python
X_train.drop(columns=['Pitchers Average Age', 'Pitchers Gender'], inplace=True)
X_test.drop(columns=['Pitchers Average Age', 'Pitchers Gender'], inplace=True)
```

```python
X_train
```

```python
industry_ohe = OneHotEncoder(sparse_output=False).set_output(transform='pandas')

industry_ohe.fit(X_train[['Industry']])

encoded_df_train = industry_ohe.transform(X_train[['Industry']])
encoded_df_test = industry_ohe.transform(X_test[['Industry']])

X_train = pd.concat([X_train, encoded_df_train], axis=1)
X_test = pd.concat([X_test, encoded_df_test], axis=1)

X_train.drop(columns=['Industry'], inplace=True)
X_test.drop(columns=['Industry'], inplace=True)
```
