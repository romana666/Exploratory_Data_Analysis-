# Data Visualization

```python
# distribution of Survived column


sns.countplot(
    data = df,
    x="Survived"
    
)
```

Insight---- The target variable (Survived) is imbalanced.

```python
# distribution of countries

df["Country"].value_counts().plot(
    kind="bar",
    figsize=(8,5)
)

plt.xticks(rotation=45)
plt.show()
```

Insight---- Country is dominated by a few categories, especially Sweden and Estonia.

```python
# distribution in Sex column 

sns.countplot(
    data = df,
    x="Sex"
    
)
```

Insight --- Number of female passengers slightly less than male passengers

```python
# distribution of age column

sns.histplot(
    data = df,
    x = "Age",
    kde = "True"
    
)

plt.show()
```

```python
#distribution in 

sns.countplot(
    data = df,
    x = "Category",
    hue = "Survived"
)
```

Insight----Category P contains significantly more survivors than Category C

```python
#bxplot - Age Distribution by Survival Status

sns.boxplot(x="Survived", y="Age", data=df)

plt.title("Age Distribution by Survival Status")
plt.show()
```

Insight-----Passengers who survived tended to be younger,while non-survivors had a higher median age and a wider age distribution
