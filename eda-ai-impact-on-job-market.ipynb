## Exploratory Data Analysis - EDA

```python
# Job_Title distribution in Industry 

plt.figure(figsize=(12, 6))

sns.countplot(
    data=df,
    x="Industry",
    hue="Job_Title",
    order=df["Industry"].value_counts().index,
    palette="tab20"
)

plt.legend(fontsize=5,title_fontsize=4)
plt.title("Job_Title Distribution in Industry")
plt.show()
```

```python
# Job_Title Distribution
plt.figure(figsize=(12,6))

sns.countplot(
    data = df,
    y = 'Job_Title',
    order=df['Job_Title'].value_counts().index
)

plt.title("Job_Title Distribution")
plt.show
```

```python
# Industry Distribution 

industry_counts = df["Industry"].value_counts()
plt.pie(
    industry_counts,
    labels=industry_counts.index,
    autopct = "%1.1f%%",
    shadow = 'True'
)
plt.title("Industry Distribution")
plt.show()
```

```python
#Country Dstribution

country_counts = df['Country'].value_counts()

plt.pie(
    country_counts,
    labels = country_counts.index,
    autopct = '%1.1f%%',
    shadow = 'True'         
)

plt.title("Country Distribution")
plt.show()
```

```python
# Education_Level Distribution

education_level_count = df['Education_Level'].value_counts()

plt.bar(
    education_level_count.index,
    education_level_count.values   
)

plt.title('Education_Level Distribution')
plt.show()
```

```python
#AI_Replacement_Risk Distribution

plt.hist(
    x=df['AI_Replacement_Risk'],
    bins=20,
    edgecolor='black'
    
)
plt.title('AI_Replacement_Risk Distribution')
plt.show()
```

```python
# Average Growth in 2030 by Upskilling

plt.figure(figsize=(6,4))

sns.barplot(
    data=df,
    x="Upskilling_Needed",
    y="Job_Growth_2030"
)

plt.title("Average Growth in 2030 by Upskilling")
plt.xlabel("Upskilling Needed")
plt.ylabel("Job Growth 2030")

plt.show()
```

```python
#Future Demand Score vs Performance Score

plt.figure(figsize=(8,6))



sns.scatterplot(
    data=df,
    x = df["Future_Demand_Score"],
    y =df["Performance_Score"],
    hue=df["Automation_Level"],
    palette="Set2"
)

plt.legend(title="Automation Level")
plt.xlabel("Future Demand Score ")
plt.ylabel("Performance Score")
plt.title("Future Demand Score vs Performance Score")

plt.show()
```

```python
#Performance Score by Hiring Trend 2026
plt.figure(figsize=(6,4))

sns.barplot(
    data=df,
    x="Hiring_Trend_2026",
    y="Performance_Score"
)

plt.title("Performance Score by Hiring Trend 2026")
plt.xlabel("Hiring Trend 2026")
plt.ylabel("Performance_Score")

plt.show()
```

```python
# Job Growth 2030 by Remote Work Possibility

plt.figure(figsize=(10,6))

sns.boxplot(
    data=df,
    x=df["Remote_Work_Possibility"],
    y=df["Job_Growth_2030"],
    hue=df["Upskilling_Needed"],
    palette="Set2",
    
)
plt.legend(title="Upskilling Needed", loc="upper right")
plt.title("Job Growth 2030 by Remote Work Possibility")
plt.xlabel("Remote Work Possibility")
plt.ylabel("Job Growth 2030")

plt.show()
```
