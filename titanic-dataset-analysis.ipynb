# Data Visualization

```python
#Survived distibution

survived = df['Survived'].value_counts()
plt.pie(
    survived,
    labels = survived.index,
    autopct = '%1.1f%%',
    shadow = True
)
plt.show
```

Insight --- In survived ditribution shows that 38.4% people survived while 61.6% people didnot survived



```python
# Age distribution 
sns.histplot(
    data = df,
    x= 'Age',
    kde= True
)
```

Insight --- The age variable is right-skewed, with most observations concentrated between 20 and 35 years.

```python
#Sex distribution 
sns.countplot(
    data =df,
    x='Sex',
    hue = 'Survived'
)

plt.show
```

Insight ---- The plot shows strong relationship between Sex and Survived.Most male passengers did not survived,whereas most female passenger survived.This indicates that gender was a significant factor influencing survival.


```python
#boxplot distribution
sns.boxplot(
    data = df,
    x = "Survived",
    y = "Fare"
)

plt.show
```

Insight---- The boxplot indicates that survivors generally paid higher ticket fares than non-survivors.The Fare variable contains many high-value outliers and shows greater variability among survivors

```python
#heatmap 
sns.heatmap(
    df.corr(numeric_only =True),
    annot =True,
    cmap = "coolwarm"
    
)
plt.show
```

Insight---- The heatmap shows that Pclass(-0.34) and Fare (0.26) have the strongest relationship with survival,indicating the passengers in higher classes and those who paid higher fares were more likely to survive.
